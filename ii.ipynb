```python
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize

f= open("C:\\Users\\ABHISHEK KUMAR\\Desktop\\r.txt")
text = f.read()
print(text)

sent = sent_tokenize(text)
print("number of sentences:",len(sent))

for i in range(len(sent)):
    print("\n sentence:",i+1,"\n",sent[i])
    
w= word_tokenize(text)
print("\n total words:",len(sent))
print(w)
```

```python
from nltk.util import ngrams
from nltk.tokenize import word_tokenize

f=open("C:\\Users\\ABHISHEK KUMAR\\Desktop\\r.txt")
text = f.read()
w= word_tokenize(text)

print(" bi-grams:\n\n",list(ngrams(w,2)))
print("\n\n tri-grams:\n\n",list(ngrams(w,3)))
```

```python
import pandas as p
import matplotlib.pyplot as m
from nltk.Probality import FreqDist

freq = FreqDist()
print("count of and word:",freq['numpy'])

freq=p.seras(dict(freq))
m.figure(figsize=(18,10))
freq.plot(kind='bar')
m.show()
```

```python
from wordcloud import WordCloud, STOPWORDS
stop_w=set(STOPWORDS)

wc=WordCloud(width=800,height=800,
             background_color='black',colormap='plasma',
             stopwords=stop_w,
             min_font_size=8).generate(text)

m.figure(figsize=(8,8),facecolor=None)
m.imshow(wc)
m.axis('off')
m.tight_layout(pad=0)
m.show()
```

```python
import numpy as np
a = np.array([1,2,3,4,5])
print(a)
s = np.sum(a)
print("sum:",s)
min = np.min(a)
print("min:",min)
max = np.max(a)
print("max:",max)
median = np.median(a)
print("median:",median)
std = np.std(a)
print("std:",std)
product = np.product(a)
print("product:",product)
```

```python
import numpy as np
import timeit
np.a=[4,5,1]
print(np.prod(np.a))
print("Time taken by vectorized product : ",end= "")
%timeit np.prod(np.a)
total = 1
for item in np.a:
 total =total*item
t = total
print(t)
print("Time taken by iterative multiplication : ",end= "")
%timeit t
```

```python
import numpy as np
import timeit
np.a=[4,5,1]
print(np.sum(np.a))
print("Time taken by vectorized sum: ",end= "")
%timeit np.sum(np.a)
total = 0
for item in np.a:
 total+=item
t = total
print(t)
print("Time taken by iterative sum : ",end= "")
%timeit t
```
